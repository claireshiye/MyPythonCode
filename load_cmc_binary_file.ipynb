```python
class BinaryStruct(Structure):
    _fields_ =[('id1', c_long),
               ('id2', c_long),
               ('rad1', c_double),
               ('rad2', c_double),
               ('m1', c_double),
               ('m2', c_double),
               ('Eint1', c_double),
               ('Eint2', c_double),
               ('a', c_double),
               ('e', c_double),
               ('inuse', c_int),
               ('bse_kw[0]', c_int),
               ('bse_kw[1]', c_int),
               ('bse_zams_mass[0]', c_double),
               ('bse_zams_mass[1]', c_double),
               ('bse_mass0[0]', c_double),
               ('bse_mass0[1]', c_double),
               ('bse_mass[0]', c_double),
               ('bse_mass[1]', c_double),
               ('bse_radius[0]', c_double),
               ('bse_radius[1]', c_double),
               ('bse_lum[0]', c_double),
               ('bse_lum[1]', c_double),
               ('bse_massc[0]', c_double),
               ('bse_massc[1]', c_double),
               ('bse_radc[0]', c_double),
               ('bse_radc[1]', c_double),
               ('bse_menv[0]', c_double),
               ('bse_menv[1]', c_double),
               ('bse_renv[0]', c_double),
               ('bse_renv[1]', c_double),
               ('bse_ospin[0]', c_double),
               ('bse_ospin[1]', c_double),
               ('bse_B_0[0]', c_double),
               ('bse_B_0[1]', c_double),
               ('bse_bacc[0]', c_double),
               ('bse_bacc[1]', c_double),
               ('bse_tacc[0]', c_double),
               ('bse_tacc[1]', c_double),
               ('bse_epoch[0]', c_double),
               ('bse_epoch[1]', c_double),
               ('bse_tms[0]', c_double),
               ('bse_tms[1]', c_double),
               ('bse_bhspin[0]', c_double),
               ('bse_bhspin[1]', c_double),
               ('bse_tphys', c_double),
               ('bse_tb', c_double),
               ('bse_bcm_dmdt[0]', c_double),
               ('bse_bcm_dmdt[1]', c_double),
               ('bse_bcm_radrol[0]', c_double),
               ('bse_bcm_radrol[1]', c_double),
               ('bse_bcm_B[0]', c_double),
               ('bse_bcm_B[1]', c_double),
               ('bse_bcm_formation[0]', c_double),
               ('bse_bcm_formation[1]', c_double),
               ('lifetime_m1', c_double),
               ('lifetime_m2', c_double),
               ('createtime_m1', c_double),
               ('createtime_m2', c_double),
               ('threebodybinary', c_int)]
```
